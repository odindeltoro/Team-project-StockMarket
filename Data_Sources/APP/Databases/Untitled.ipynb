```python
import pandas as pd
from sqlalchemy import create_engine
from postgresql_key import password
```

```python
financials=pd.read_json("financialsNASDAQ.json")
historical=pd.read_json("historicalStockPriceNASDAQ.json")
quotes=pd.read_json("quotesStockPriceNASDAQ.json")
```

```python
financials.columns= financials.columns.str.lower()
historical.columns= historical.columns.str.lower()
quotes.columns= quotes.columns.str.lower()
```

```python
quotes
```

```python
engine= create_engine(f"postgresql+psycopg2://postgres:{password}@/StockPrices")
#Send appending without index
financials.to_sql("financials", if_exists="append",index=False, con= engine)
historical.to_sql("historical", if_exists="append",index=False, con= engine)
quotes.to_sql("quotes", if_exists="append",index=False, con= engine)
engine.dispose()
```
