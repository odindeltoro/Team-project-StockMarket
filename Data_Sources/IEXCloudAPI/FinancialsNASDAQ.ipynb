```python
# Import dependencies
import requests
import json
import pandas as pd
import numpy as np
import os
```

```python
# Files to read
financialsNASDAQ = './financials_NASDAQ.csv'
catalogueNASDAQ = './Databases/nasdaq.csv'
# Read files
financialsNASDAQ_df = pd.read_csv(financialsNASDAQ, encoding='utf-8')
catalogueNASDAQ_df = pd.read_csv(catalogueNASDAQ, encoding='utf-8')
```

```python
catalogueNASDAQ_df
```

```python
catalogueNASDAQ_df = catalogueNASDAQ_df[['Ticker','Name','Industry']]
catalogueNASDAQ_df= catalogueNASDAQ_df.rename(columns={'Ticker': 'Symbol'})
catalogueNASDAQ_df
```

```python
financialsNASDAQ_df
```

```python
financialsNASDAQ_df = financialsNASDAQ_df.rename({'symbol': 'Symbol', 'reportDate': 'ReportDate', 
                    'fiscalDate': 'FiscalDate','currency': 'Currency',
                    'grossProfit': 'GrossProfit','costOfRevenue': 'CostofRevenue',
                    'operatingRevenue': 'OperatingRevenue','totalRevenue': 'TotalRevenue',
                    'operatingIncome': 'OperatingIncome','netIncome': 'NetIncome',
                    'researchAndDevelopment': 'ResearchandDevelopment','operatingExpense': 'OperatingExpense',
                    'currentAssets': 'CurrentAssets','totalAssets': 'TotalAssets',
                    'totalLiabilities': 'TotalLiabilities','currentCash': 'CurrentCash',
                    'currentDebt': 'CurrentDebt','shortTermDebt': 'ShortTermDebt',
                    'longTermDebt': 'LongTermDebt','totalCash': 'TotalCash',
                    'totalDebt': 'TotalDebt','shareholderEquity': 'ShareholderEquity',
                    'cashChange': 'CashChange','cashFlow': 'CashFlow'}, axis=1)
financialsNASDAQ_df
```

```python
financialsNASDAQMerge_df = pd.merge(financialsNASDAQ_df,catalogueNASDAQ_df,on='Symbol',how='outer')
financialsNASDAQMerge_df
```

```python
# Create csv and json file for database
financialsNASDAQMerge_df.to_csv('Databases/financialsNASDAQ.csv')
financialsNASDAQMerge_df.to_json('Databases/financialsNASDAQ.json',orient = 'records')
```
