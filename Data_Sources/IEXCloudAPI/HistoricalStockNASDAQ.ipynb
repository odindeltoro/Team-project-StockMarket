```python
# Import dependencies
import requests
import json
import pandas as pd
import numpy as np
import os
```

```python
# Files to read
historicalNASDAQ = './historical_stock_NASDAQ.csv'
catalogueNASDAQ = './Databases/nasdaq.csv'
# Read files
historicalNASDAQ_df = pd.read_csv(historicalNASDAQ, encoding='utf-8')
catalogueNASDAQ_df = pd.read_csv(catalogueNASDAQ, encoding='utf-8')
```

```python
catalogueNASDAQ_df
```

```python
catalogueNASDAQ_df = catalogueNASDAQ_df[['Ticker','Name','Industry']]
catalogueNASDAQ_df= catalogueNASDAQ_df.rename(columns={'Ticker': 'Symbol'})
catalogueNASDAQ_df
```

```python
# del historicalNASDAQ_df["currency"]
historicalNASDAQ_df = historicalNASDAQ_df.rename({'symbol': 'Symbol', 'date': 'Date', 
                    'uClose': 'UnadjustedClosePrice','uOpen': 'UnadjustedOpenPrice',
                    'uHigh': 'UnadjustedHighPrice','uLow': 'UnadjustedLowPrice',
                    'uVolume': 'UnadjustedVolume','close': 'AdjustedClosePrice',
                    'open': 'AdjustedOpenPrice','high': 'AdjustedHighPrice',
                    'low': 'AdjustedLowPrice','volume': 'AdjustedVolume',
                    'change': 'ChangefromPreviousDay','changePercent': 'ChangePercentagefromPreviousDay',
                    'label': 'Date2','changeOverTime': 'ChangePercentagefromFirstValue'}, axis=1)
# historicalNASDAQ_df = catalogueNASDAQ_df[['Symbol','Date','Date2','UnadjustedClosePrice','UnadjustedOpenPrice',
#                     'UnadjustedHighPrice','UnadjustedLowPrice','UnadjustedVolume','AdjustedClosePrice',
#                     'AdjustedOpenPrice','AdjustedHighPrice','AdjustedLowPrice','AdjustedVolume',
#                     'ChangefromPreviousDay','ChangePercentagefromPreviousDay','ChangePercentagefromFirstValue']]
historicalNASDAQ_df
```

```python
historicalNASDAQMerge_df = pd.merge(historicalNASDAQ_df,catalogueNASDAQ_df,on='Symbol',how='outer')
historicalNASDAQMerge_df
```

```python
# Create csv and json file for database
historicalNASDAQMerge_df.to_csv('Databases/historicalStockPriceNASDAQ.csv')
historicalNASDAQMerge_df.to_json('Databases/historicalStockPriceNASDAQ.json',orient = 'records')
```
