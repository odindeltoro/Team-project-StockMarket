```python
# URL for testing
base_url_sandbox = "https://sandbox.iexapis.com/" 
version = "stable/stock/"
endpoint = "chart/"
token_sandbox = apiKeySandbox
```

```python
# symbol_quote = "AAPL/"
# # Extract data for quotes
# HTTP_request_quotes = f"{base_url_sandbox}{version}{symbol_quote}quote?token={token_sandbox}"
# results_sandbox_quotes = requests.get(HTTP_request_quotes).json()
# c=pd.DataFrame(results_sandbox_quotes,index=[0]).columns
# a=pd.DataFrame()
```

```python
full=[]
for x in tickers:
    symbol_quote = x+"/"
    # Extract data for quotes
    HTTP_request_quotes = f"{base_url_sandbox}{version}{symbol_quote}quote?token={token_sandbox}"
    results_sandbox_quotes = requests.get(HTTP_request_quotes).json()
    c=results_sandbox_quotes
    full.append(c)
```

```python
full=pd.DataFrame(full)
full=full[["symbol","open","close","high","low","previousClose","changePercent","latestTime","volume","avgTotalVolume","week52High","week52Low"]]
full
```

```python
full = full.rename({'symbol': 'Symbol', 'open': 'OpenPrice', 
                    'close': 'ClosePrice','high': 'HighPrice',
                    'low': 'LowPrice','previousClose': 'PreviousDayClosePrice',
                    'changePercent': 'ChangePercentagefromPreviousDay','latestTime': 'LatestTime',
                    'volume': 'Volume','avgTotalVolume': '30DayAverageVolume',
                    'week52High': 'Adjusted52WeekHigh','week52Low': 'Adjusted52WeekLow'}, axis=1)
full
```

```python
# Files to read
catalogueNASDAQ = './nasdaq.csv'
# Read files
catalogueNASDAQ_df = pd.read_csv(catalogueNASDAQ, encoding='utf-8')
```

```python
catalogueNASDAQ_df = catalogueNASDAQ_df[['Ticker','Name','Sector','Industry']]
catalogueNASDAQ_df= catalogueNASDAQ_df.rename(columns={'Ticker': 'Symbol'})
catalogueNASDAQ_df
```

```python
quotesNASDAQMerge_df = pd.merge(full,catalogueNASDAQ_df,on='Symbol',how='outer')
quotesNASDAQMerge_df
```

```python
# Create csv and json file for database
quotesNASDAQMerge_df.to_csv('Databases/quotesStockPriceNASDAQ.csv')
quotesNASDAQMerge_df.to_json('Databases/quotesStockPriceNASDAQ.json',orient = 'records')
```
