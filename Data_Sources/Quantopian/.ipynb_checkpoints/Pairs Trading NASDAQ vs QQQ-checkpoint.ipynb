```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
start = '31-08-2015'
end = '30-08-2020'
QQQ = get_pricing('QQQ',start_date=start,end_date=end)
NASDAQ = get_pricing('NDAQ',start_date=start,end_date=end)
```

```python
QQQ['close_price'].plot(label='QQQ',figsize=(12,8),color='dodgerblue')
NASDAQ['close_price'].plot(label='NASDAQ',color='tab:orange')
plt.grid(b=None)
plt.xlabel('Dates',style='italic',color='black',family='Times New Roman',fontsize=10)
plt.ylabel('Adjusted Closed Price USD',style='italic',color='black',family='Times New Roman',fontsize=10)
plt.title('5Y NASDAQ vs QQQ',fontsize=25,color='steelblue',fontweight='semibold', family='Times New Roman')
plt.legend()
```

```python
np.corrcoef(NASDAQ['close_price'],QQQ['close_price'])
```

```python
spread = NASDAQ['close_price'] - QQQ['close_price']
spread.plot(label='Spread',figsize=(12,8),color='dodgerblue')
plt.axhline(spread.mean(),c='firebrick')
plt.grid(b=None)
plt.xlabel('Dates',style='italic',color='black',family='Times New Roman',fontsize=10)
plt.ylabel('Adjusted Closed Price USD',style='italic',color='black',family='Times New Roman',fontsize=10)
plt.title('5Y NASDAQ vs QQQ Spread (High vs Low)',fontsize=25,color='steelblue',fontweight='semibold', family='Times New Roman')
plt.legend()
```

```python
def zscore(stocks):
    return (stocks - stocks.mean()) / np.std(stocks)
```

```python
zscore(spread).plot(figsize=(14,8), color='dodgerblue')
plt.axhline(zscore(spread).mean(),color='black')
plt.axhline(1.0,c='teal',ls='--')
plt.axhline(-1.0,c='firebrick',ls='--')
plt.grid(b=None)
plt.xlabel('Dates',style='italic',color='black',family='Times New Roman',fontsize=10)
plt.ylabel('Adjusted Closed Price USD',style='italic',color='black',family='Times New Roman',fontsize=10)
plt.title('5Y NASDAQ vs QQQ Normalized Spread',fontsize=25,color='steelblue',fontweight='semibold', family='Times New Roman')
plt.legend()
```

```python
spread_mavg1 = spread.rolling(1).mean()
spread_mavg30 = spread.rolling(30).mean()
std_30 = spread.rolling(30).std()
zscore_30_1 = (spread_mavg1 - spread_mavg30)/std_30
```

```python
zscore_30_1.plot(figsize=(12,8),label='Rolling 30 day Z score',color='dodgerblue')
plt.axhline(0, color='black')
plt.axhline(1.0, color='firebrick', linestyle='--');
plt.grid(b=None)
plt.xlabel('Dates',style='italic',color='black',family='Times New Roman',fontsize=10)
plt.ylabel('Adjusted Closed Price USD',style='italic',color='black',family='Times New Roman',fontsize=10)
plt.title('5Y NASDAQ vs QQQ Rolling 30 Day Z score',fontsize=25,color='steelblue',fontweight='semibold', family='Times New Roman')
plt.legend()
```
