# Linear analysis and K Nearest Neighbor

```python
# Import dependencies for stock analysis
import pandas as pd
import datetime
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import style
from pandas.plotting import scatter_matrix
import seaborn as sns
import math
import numpy as np
from sklearn import preprocessing, svm
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import and read data from historical stock prices database
stock_data = pd.read_csv("../Data_Sources/IEXCloudAPI/Databases/historicalStockPriceNASDAQ.csv")
del stock_data["Unnamed: 0"]
stock_data.head()
```

```python
# Change date column to datetime
stock_data["Date"] = stock_data["Date"].apply(pd.to_datetime)
stock_data.head()
```

```python
# Chosen companies
chosen_company = "AAPL"
chosen_company2 = "MSFT"
chosen_company3 = "FB"
chosen_company4 = "INTC"
chosen_company5 = "CSCO"
company_df = stock_data[stock_data["Symbol"]==chosen_company]
company2_df = stock_data[stock_data["Symbol"]==chosen_company2]
company3_df = stock_data[stock_data["Symbol"]==chosen_company3]
company4_df = stock_data[stock_data["Symbol"]==chosen_company4]
company5_df = stock_data[stock_data["Symbol"]==chosen_company5]
company_df.head()
```

### Rolling mean

```python
# Moving Average for last 365 days to reduce noise in data with close price and make line smoother and determine trend (increase or decrease)
close_price = company_df["AdjustedClosePrice"]
mavg = close_price.rolling(window = 365).mean()
mavg.tail()
```

```python
# Plot results
plt.figure(figsize=(30,20))
plt.plot(company_df["Date"], close_price, ms=10,color="dodgerblue",linewidth=1.5, label=company_df["Name"][0])
plt.plot(company_df["Date"], mavg, ms=10,color="tab:orange",linewidth=1.5, label = "Moving Average")
plt.xlabel("Year",style="italic",color="black",alpha=0.8,family="Times New Roman",fontsize=40)
plt.ylabel("Close price USD",style="italic",color="black",alpha=0.8,family="Times New Roman",fontsize=40)
plt.title("1Y Moving Average",fontsize=70,color="black",alpha=0.8,fontweight="semibold", family="Times New Roman")
plt.xticks(color="black",alpha=0.8,fontsize=25)
plt.yticks(color="black",alpha=0.8,fontsize=25)
plt.legend(loc="upper left",prop={"size": 30,"weight":"bold"})
plt.savefig('Images/MovingAverage.png')
plt.show()
```

### Return Deviation

```python
# Determine the probability distribution of investments returns, the more stable the less riskier investment
investment_return = close_price / close_price.shift(1)-1
plt.figure(figsize=(30,20))
plt.plot(company_df["Date"], investment_return, ms=10,color="dodgerblue",linewidth=1.5, label = "Return")
plt.xlabel("Year",style="italic",color="black",alpha=0.8,family="Times New Roman",fontsize=40)
plt.ylabel("Probability",style="italic",color="black",alpha=0.8,family="Times New Roman",fontsize=40)
plt.title("Investment Return",fontsize=70,color="black",alpha=0.8,fontweight="semibold", family="Times New Roman")
plt.xticks(color="black",alpha=0.8,fontsize=25)
plt.yticks(color="black",alpha=0.8,fontsize=25)
plt.legend(loc="upper left",prop={"size": 30,"weight":"bold"})
plt.savefig('Images/InvestmentReturnCompany.png')
plt.show()
```

### Predict price for chosen company with ML Linear Analysis, QDA and KNN

```python
# ML Simple Linear Analysis, Quadratic Discriminant Analysis and K Nearest Neighbor, calculate High Low percentage and Percentage Change
regression = company_df.loc[:,["Date", "AdjustedClosePrice","AdjustedVolume"]]
regression["HighLow_percentage"] = (company_df["AdjustedHighPrice"] - company_df["AdjustedLowPrice"]) / company_df["AdjustedClosePrice"] * 100.0
regression["Percentage_change"] = (company_df["AdjustedClosePrice"] - company_df["AdjustedOpenPrice"]) / company_df["AdjustedOpenPrice"] * 100.0
regression = regression.set_index("Date")
regression.head()
```

```python
# Pre processing and Cross Validation
# Drop missing value
regression.fillna(value=-99999, inplace=True)
print(regression.shape)
# Separate one percent of the data to forecast
forecast_out = int(math.ceil(0.01 * len(regression)))
# Separating the label here and predict the Adjusted Close Price
forecast_col = "AdjustedClosePrice"
regression["Label"] = regression[forecast_col].shift(-forecast_out)
X = np.array(regression.drop(["Label"], 1))
# Scale the X for everyone to have the same distribution for linear regression
X = preprocessing.scale(X)
# Finally find Data Series of late X and early X (train) for model generation and evaluation
X_lately = X[-forecast_out:]
X = X[:-forecast_out]
# Separate label and identify it as y
y = np.array(regression["Label"])
y = y[:-forecast_out]
print("Dimension of X",X.shape)
print("Dimension of y",y.shape)
X
```

```python
# Separating values train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

### Model Algorithm

```python
# Linear regression
classificationRegression = LinearRegression(n_jobs=-1)
classificationRegression.fit(X_train, y_train)
# Quadratic Regression 2
classificationPolynomial2 = make_pipeline(PolynomialFeatures(2), Ridge())
classificationPolynomial2.fit(X_train, y_train)
# Quadratic Regression 3
classificationPolynomial3 = make_pipeline(PolynomialFeatures(3), Ridge())
classificationPolynomial3.fit(X_train, y_train)
# KNN Regression
classificationKNN = KNeighborsRegressor(n_neighbors=2)
classificationKNN.fit(X_train, y_train)
```

```python
# Testing the model with score method to find the accuracy
confidenceRegression = classificationRegression.score(X_test, y_test)
confidencePolynomial2 = classificationPolynomial2.score(X_test,y_test)
confidencePolynomial3 = classificationPolynomial3.score(X_test,y_test)
confidenceKNN = classificationKNN.score(X_test, y_test)
print("The linear regression confidence is ",confidenceRegression)
print("The quadratic regression 2 confidence is ",confidencePolynomial2)
print("The quadratic regression 3 confidence is ",confidencePolynomial3)
print("The knn regression confidence is ",confidenceKNN)
```

The most accurate is the linear regression

### Prediction

```python
last_date = regression.iloc[-1].name
last_unix = last_date
next_unix = last_unix + datetime.timedelta(days=1)
for i in forecast_set:
    next_date = next_unix
    next_unix += datetime.timedelta(days=1)
    regression.loc[next_date] = [np.nan for _ in range(len(regression.columns)-1)]+[i]
```
